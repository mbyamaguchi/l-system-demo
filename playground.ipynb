# L-System Playground

## 形式的定義

- アルファベット：使用する記号の有限集合

- 公理（初期文字列）：成長の出発点となる文字列

- 規則（置換規則）：各記号を別の文字列に写す規則

- 生成規則：各ステップで、文字列中のすべての記号に同時に規則を適用する

```python
# Definition
Alphabet = {"A", "B"}
Rules = {
    "A": "AB",
    "B": "A",
}
Axiom = "A"
```

```python
# Growth

def grow(axiom, rules, steps):
    for _ in range(steps):
        new_chars = []
        for char in axiom:
            new_chars.append(rules[char])
        axiom = new_chars
    return "".join(axiom)
```

```python
gen10 = grow(Axiom, Rules, 10)
```
